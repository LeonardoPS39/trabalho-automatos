```python
import re

def verificar_string(texto):
    padrao = r'^[AB\s]+$'

    if(re.fullmatch(padrao, texto)):
        return True
    else:
        return False

# Testando a função
print(verificar_string("AAB "))  # Deve retornar True
print(verificar_string("AaB "))  # Deve retornar False
```

**Descrição do Programa:** Este programa define uma função chamada verificar_string que aceita uma string como entrada. A função usa uma expressão regular para verificar se a string corresponde ao padrão a*b(a+b). Se a string corresponder ao padrão, a função retorna True, caso contrário, retorna False.

**Autômato Finito Não-Determinístico com Movimentos Vazios (AFNe):** A expressão regular usada neste programa pode ser representada por um AFNe. 


**Expressões Regulares na Notação Algébrica:** A expressão regular usada neste programa é ^a\*b[ab]\$. Esta expressão regular corresponde a qualquer sequência de caracteres que comece (^) com zero ou mais (*) 'a’s (a), seguido por um ‘b’ (b), e terminando ($) com um ‘a’ ou um ‘b’ ([ab]).
